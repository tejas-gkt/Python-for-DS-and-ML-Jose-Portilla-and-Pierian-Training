```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
```

```python
plt.plot(1,2,2)
plt.plot(y,x,'*b')
```

```python
#object oriented plotting
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("title")
#plt.show()
```
